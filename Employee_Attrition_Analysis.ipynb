```python
import scipy.stats as stats
```

```python
# Attrition rate by OverTime
ot_table = pd.crosstab(df["OverTime"], df["Attrition"], normalize="index") * 100
print(ot_table)
```

```python
ot_table.plot(kind="bar", stacked=True)
plt.title("Attrition Rate: OverTime vs No OverTime")
plt.ylabel("Percentage")
plt.show()
```

```python
contingency_table = pd.crosstab(df["OverTime"], df["Attrition"])

chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print("Chi-square:", chi2)
print("p-value:", p)
```

```python
if p < 0.05:
    print("Statistically significant relationship between OverTime and Attrition")
else:
    print("No statistically significant relationship found")
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
df_encoded = df.copy()

df_encoded["Attrition"] = df_encoded["Attrition"].map({"Yes": 1, "No": 0})

df_encoded = pd.get_dummies(df_encoded, drop_first=True)

X = df_encoded.drop("Attrition", axis=1)
y = df_encoded["Attrition"]
```

```python
rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X, y)
```

```python
importances = pd.Series(rf.feature_importances_, index=X.columns)
top_features = importances.sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_features.plot(kind="bar")
plt.title("Top Features Influencing Attrition (EDA)")
plt.show()
```

## Hypothesis 1
